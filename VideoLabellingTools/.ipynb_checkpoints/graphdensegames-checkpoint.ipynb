```python
#!/usr/bin/env python3
import csv
import re
from pathlib import Path
import os
import matplotlib.pyplot as plt

# --- Configuration ---
LABELS_DIR = Path("/opt/dlami/nvme/opentt_data/CreateOpenTTScoreLabels/VideoLabellingTools/labels/cleaned_scores")
DENSE_DATA_DIR = Path("/opt/dlami/nvme/opentt_data/CreateOpenTTScoreLabels/VideoLabellingTools/labels/densified_scores")
# --- End Configuration ---

SCORE_RX = re.compile(r"^\s*(\d+)\s*-\s*(\d+)\s*$")

def parse_score(cell: str):
    """Parse score string like '3-2' into (3, 2)."""
    if not cell:
        return 0, 0
    m = SCORE_RX.match(cell)
    return (int(m.group(1)), int(m.group(2))) if m else (0, 0)

def read_csv_scores(path: Path):
    """Read 'frame,score' CSV and return lists: frames, p1_scores, p2_scores."""
    frames, p1s, p2s = [], [], []
    if not path.exists():
        return frames, p1s, p2s
    with path.open(newline="") as f:
        reader = csv.reader(f)
        header = next(reader, None)
        for row in reader:
            if not row:
                continue
            try:
                frame = int(row[0])
                score_str = row[1].strip() if len(row) > 1 else ""
                p1, p2 = parse_score(score_str)
                frames.append(frame)
                p1s.append(p1)
                p2s.append(p2)
            except Exception:
                continue

    # Sort by frame
    if not frames:
        return [], [], []
    order = sorted(range(len(frames)), key=lambda i: frames[i])
    return [frames[i] for i in order], [p1s[i] for i in order], [p2s[i] for i in order]

def plot_game(filename):
    sparse_path = LABELS_DIR / filename
    dense_path = DENSE_DATA_DIR / filename

    sparse_frames, sparse_p1, sparse_p2 = read_csv_scores(sparse_path)
    dense_frames, dense_p1, dense_p2 = read_csv_scores(dense_path)

    if not sparse_frames or not dense_frames:
        print(f"⚠️ Skipping {filename}: sparse or dense data not found.")
        return

    fig, axes = plt.subplots(1, 2, figsize=(16, 6), sharey=True)
    fig.suptitle(f"{filename} — Sparse vs Dense Labels", fontsize=16)

    # Sparse
    axes[0].step(sparse_frames, sparse_p1, where='post', linestyle="--", label="Left Player", color="blue")
    axes[0].step(sparse_frames, sparse_p2, where='post', linestyle="--", label="Right Player", color="red")
    axes[0].set_title("Sparse Labels")
    axes[0].set_xlabel("Frame ID")
    axes[0].set_ylabel("Score")
    axes[0].legend()
    axes[0].grid(True, alpha=0.3)

    # Dense
    axes[1].step(dense_frames, dense_p1, where='post', label="Left Player", color="blue")
    axes[1].step(dense_frames, dense_p2, where='post', label="Right Player", color="red")
    axes[1].set_title("Dense Labels (Forward-Fill)")
    axes[1].set_xlabel("Frame ID")
    axes[1].legend()
    axes[1].grid(True, alpha=0.3)

    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()

def main():
    for file in os.scandir(LABELS_DIR):
        if file.is_file() and file.name.endswith(".csv"):
            plot_game(file.name)

if __name__ == "__main__":
    main()
```

```python
#!/usr/bin/env python3
"""
plot_spliced_game.py

Plots score progression for spliced game CSVs (e.g., game_1a_scores.csv, game_1b_scores.csv).
"""

import csv
import matplotlib.pyplot as plt
from pathlib import Path

def read_csv_scores(csv_path: Path):
    frames, left, right = [], [], []
    with open(csv_path, newline="") as f:
        reader = csv.DictReader(f)
        for row in reader:
            fr = int(row["frame"])
            score_str = row.get("score", "").strip()
            if "-" not in score_str:
                continue
            try:
                l, r = map(int, score_str.split("-"))
            except ValueError:
                continue
            frames.append(fr)
            left.append(l)
            right.append(r)
    return frames, left, right

def plot_two_parts(csv_a, csv_b, title="Spliced Game"):
    f1, l1, r1 = read_csv_scores(csv_a)
    f2, l2, r2 = read_csv_scores(csv_b)

    fig, axes = plt.subplots(1, 2, figsize=(14, 5), sharey=True)
    fig.suptitle(title, fontsize=16)

    # Part A
    axes[0].plot(f1, l1, label="Left Player", color="blue", linewidth=2)
    axes[0].plot(f1, r1, label="Right Player", color="red", linewidth=2)
    axes[0].set_title(f"{csv_a.stem}")
    axes[0].set_xlabel("Frame")
    axes[0].set_ylabel("Score")
    axes[0].grid(True, alpha=0.3)
    axes[0].legend()

    # Part B
    axes[1].plot(f2, l2, label="Left Player", color="blue", linewidth=2)
    axes[1].plot(f2, r2, label="Right Player", color="red", linewidth=2)
    axes[1].set_title(f"{csv_b.stem}")
    axes[1].set_xlabel("Frame")
    axes[1].grid(True, alpha=0.3)
    axes[1].legend()

    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    base = Path("/opt/dlami/nvme/opentt_data/CreateOpenTTScoreLabels/VideoLabellingTools/labels/densified_scores")
    csv_a = base / "game_1a_scores.csv"
    csv_b = base / "game_1b_scores.csv"
    plot_two_parts(csv_a, csv_b, title="Game 1 Spliced")
```

```python
#!/usr/bin/env python3
"""
overwrite_scores.py

Directly overwrite scores in a CSV for a given frame range. This is because some of the scores flip because players switch sides.
"""

import csv
from pathlib import Path

csv_path = Path("/opt/dlami/nvme/opentt_data/CreateOpenTTScoreLabels/VideoLabellingTools/labels/densified_scores/game_2b_scores.csv")

start_frame = 149900
end_frame   = 150500
new_score   = "5-3"

rows = []
with open(csv_path, newline="") as f:
    reader = csv.DictReader(f)
    for row in reader:
        fr = int(row["frame"])
        if start_frame <= fr <= end_frame:
            row["score"] = new_score
        rows.append(row)

with open(csv_path, "w", newline="") as f:
    writer = csv.DictWriter(f, fieldnames=["frame", "score"])
    writer.writeheader()
    writer.writerows(rows)

print(f"✅ Updated {csv_path} (frames {start_frame}-{end_frame} set to {new_score})")
```
