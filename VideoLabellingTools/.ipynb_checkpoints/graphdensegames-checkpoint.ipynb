```python
#!/usr/bin/env python3
import csv
import re
from pathlib import Path
import os

import matplotlib.pyplot as plt

# Adjusted to a more generic name for clarity
LABELS_DIR = Path("/opt/dlami/nvme/opentt_data/CreateOpenTTScoreLabels/VideoLabellingTools/labels/cleaned_scores")
DENSE_DATA_DIR = Path("/opt/dlami/nvme/opentt_data/CreateOpenTTScoreLabels/VideoLabellingTools/labels/densified_scores")


SCORE_RX = re.compile(r"^\s*(\d+)\s*-\s*(\d+)\s*$")

def parse_score(cell: str):
    """Return (p1, p2) from 'N-M'; blanks/invalid -> (0,0)."""
    if not cell:
        return 0, 0
    m = SCORE_RX.match(cell)
    if not m:
        return 0, 0
    return int(m.group(1)), int(m.group(2))

def read_csv_scores(path: Path):
    """Return lists x=frames, y1=P1, y2=P2 from a labels CSV."""
    frames, p1s, p2s = [], [], []
    if not path.exists():
        return frames, p1s, p2s
    with path.open(newline="") as f:
        reader = csv.reader(f)
        header = next(reader, None)  # ['frame','score']
        for row in reader:
            if not row:
                continue
            try:
                frame = int(row[0])
            except Exception:
                continue
            p1, p2 = parse_score(row[1].strip() if len(row) > 1 else "")
            frames.append(frame)
            p1s.append(p1)
            p2s.append(p2)
    # sort by frame (just in case)
    if not frames:
        return [], [], []
    order = sorted(range(len(frames)), key=lambda i: frames[i])
    frames = [frames[i] for i in order]
    p1s = [p1s[i] for i in order]
    p2s = [p2s[i] for i in order]
    return frames, p1s, p2s

def plot_game(file):
    game_path = f"{file}"

    full_game_path = LABELS_DIR / game_path
    dense_game_path = DENSE_DATA_DIR / game_path

    # Read sparse labels
    frames, left_player, right_player = read_csv_scores(full_game_path)
    # Read dense labels
    dense_frames, dleft_player, dright_player = read_csv_scores(dense_game_path)
    
    if not frames or not dense_frames:
        print(f"Skipping {file}, no data found.")
        return

    fig, axes = plt.subplots(1, 2, figsize=(16, 6), sharey=True)
    fig.suptitle(f"{file} — Sparse vs Dense OCR labels", fontsize=16)

    # Sparse data
    # Using step() here too for consistency, it works well for sparse data.
    axes[0].step(frames, left_player, where='post', linestyle="--", label="Left Player", linewidth=2, color="blue")
    axes[0].step(frames, right_player, where='post', linestyle="--", label="Right Player", linewidth=2, color="red")
    axes[0].set_title("Sparse Labels")
    axes[0].set_xlabel("Frame ID")
    axes[0].set_ylabel("Score")
    axes[0].grid(True, alpha=0.3)
    axes[0].legend()

    # Dense data
    # Use step() instead of plot() for a clean step chart
    axes[1].step(dense_frames, dleft_player, where='post', label="Left Player", linewidth=1.5, color="blue")
    axes[1].step(dense_frames, dright_player, where='post', label="Right Player", linewidth=1.5, color="red")
    axes[1].set_title("Dense Labels")
    axes[1].set_xlabel("Frame ID")
    axes[1].grid(True, alpha=0.3)
    axes[1].legend()

    plt.tight_layout(rect=[0, 0.03, 1, 0.95]) # Adjust layout to prevent title overlap
    plt.show()


def main():
    for file in os.scandir(LABELS_DIR):
        if file.is_file() and file.name.endswith('.csv'):
            plot_game(file.name)

if __name__ == "__main__":
    main()
```
